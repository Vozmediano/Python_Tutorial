# 3.- INDICES Y SLICING

En Python, los índices y el "slicing" (cortar o rebanar) son conceptos utilizados para acceder y manipular partes de secuencias como cadenas de texto (strings), listas, y tuplas.

Índices:

Un índice es un número que indica la posición de un elemento en una secuencia. En Python, los índices comienzan en 0. Por lo tanto, el primer elemento tiene el índice 0, el segundo elemento tiene el índice 1, y así sucesivamente.
También puedes usar índices negativos para acceder a los elementos desde el final de la secuencia. Por ejemplo, -1 se refiere al último elemento, -2 al penúltimo, y así sucesivamente.
Slicing (Rebanado):

El slicing se utiliza para obtener una subsecuencia o "rebanada" de una secuencia. Se especifica mediante un rango de índices, usando la sintaxis secuencia[inicio:fin], donde inicio es el índice donde comienza el corte, y fin es el índice donde termina (sin incluir el elemento en este índice final).
También se puede especificar un tercer parámetro, paso, como en secuencia[inicio:fin:paso], que determina el intervalo de los elementos seleccionados. Por ejemplo, un paso de 2 toma cada segundo elemento.
Si se omite inicio, el slicing comienza desde el principio de la secuencia. Si se omite fin, se incluyen todos los elementos hasta el final de la secuencia.

```python
# Ejemplo:

# Considera la lista lista = [0, 1, 2, 3, 4, 5]:
# lista[2] devuelve 2 (elemento en el índice 2).
# lista[-1] devuelve 5 (el último elemento).
# lista[1:4] devuelve [1, 2, 3].
# lista[:3] devuelve [0, 1, 2] (desde el principio hasta el índice 3, sin incluirlo).
# lista[2:] devuelve [2, 3, 4, 5] (desde el índice 2 hasta el final).
# lista[::2] devuelve [0, 2, 4] (todos los elementos con índices pares).
```

# 4.- FUNCIONES MATEMATICAS

Python incluye un conjunto de funciones matemáticas incorporadas en el módulo math, así como algunas funciones matemáticas básicas disponibles directamente en el lenguaje. Aquí están algunas de las más comunes:

Funciones Básicas (sin necesidad de importar math):

- abs(x): Devuelve el valor absoluto de x.
- round(x[, n]): Redondea x al n número de decimales (si se proporciona n). Si no se proporciona n, redondea a un entero.
- min(x1, x2, ..., xn): Devuelve el menor de los valores proporcionados.
- max(x1, x2, ..., xn): Devuelve el mayor de los valores proporcionados.
- sum(iterable[, start]): Suma los elementos de un iterable, empezando por start si se proporciona.

Funciones del Módulo math:
Para usar estas funciones, primero debes importar el módulo con import math.

- math.sqrt(x): Calcula la raíz cuadrada de x.
- math.pow(x, y): Eleva x a la potencia de y.
- math.exp(x): Calcula e^x, donde e es la base de los logaritmos naturales.
- math.log(x[, base]): Calcula el logaritmo de x con la base especificada; si no se especifica la base, se utiliza el logaritmo natural.
- math.sin(x), math.cos(x), math.tan(x): Funciones trigonométricas que calculan el seno, coseno y tangente de x (donde x está en radianes).
- math.radians(x): Convierte x de grados a radianes.
- math.degrees(x): Convierte x de radianes a grados.
- math.factorial(x): Calcula el factorial de x.

# 7.- CONDICIONALES: IF, ELIF y ELSE

Los condicionales en Python son estructuras de control de flujo que permiten ejecutar diferentes bloques de código dependiendo de si una condición específica es verdadera (True) o falsa (False). Estas estructuras son fundamentales en la programación, ya que permiten tomar decisiones dentro de un programa, controlando así su comportamiento basado en ciertas condiciones.

Los principales condicionales en Python son if, elif (abreviatura de "else if"), y else:

- if: Evalúa si una condición es verdadera. Si lo es, ejecuta el bloque de código indentado que le sigue.

- elif: Se utiliza después de un if para evaluar otra condición si la condición anterior no fue verdadera. Puedes tener varios elif después de un if.

- else: Se coloca al final de una cadena de if y elif, y se ejecuta si ninguna de las condiciones anteriores fue verdadera. else no lleva una condición.

```python
num1= int(input("introduce un numero: "))
num2= int(input("introduce un numero: "))

if num1==num2:
  print("los numeros son iguales")
else:
  print("son diferentes")
```

# 12.- CICLOS FOR

```python
# Los ciclos for en Python son estructuras de control que se utilizan para iterar
# sobre una secuencia de elementos. Esto puede incluir listas, tuplas,
# diccionarios, sets, strings o cualquier objeto iterable. El bucle for permite
# ejecutar un bloque de código repetidamente, una vez por cada elemento en la
# secuencia.

# La sintaxis básica de un bucle for en Python es:

for elemento in secuencia:
    # Hacer algo con 'elemento'

# Donde elemento es la variable que toma el valor del elemento actual en la
# iteración, y secuencia es la colección de elementos a iterar.

# Por ejemplo, si tienes una lista de frutas, puedes usar un bucle for para
# imprimir el nombre de cada fruta:

frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
  print(fruta)

# Este bucle se repetirá tres veces, ya que hay tres elementos en la lista frutas.
# En cada iteración, la variable fruta tomará el valor de uno de los elementos de
# la lista ("manzana", luego "banana", y finalmente "cereza"), y ejecutará el
# bloque de código dentro del bucle, que en este caso es simplemente imprimir
# el nombre de la fruta.

# Los bucles for son una herramienta fundamental en Python que facilitan la
# realización de tareas repetitivas sobre colecciones de datos.
```

```python
# La comprensión de listas (list comprehension) en Python es una forma concisa y
# conveniente de crear listas. Esencialmente, es una sintaxis que permite
# representar operaciones que involucran listas de una manera que es más compacta
# y a menudo más legible que usar bucles for convencionales.

# La sintaxis básica de una comprensión de lista es:

expresion for elemento in iterable if condicion

# expresión: Es lo que se va a hacer con cada elemento del iterable, por ejemplo,
# podría ser una operación matemática o una llamada a función.
# elemento: Es el nombre de la variable que toma el valor del elemento actual en
# la iteración.
# iterable: Es una colección de elementos que puede ser una lista, un rango, un
# string, etc.
# condición: Es opcional y se usa para incluir en la nueva lista solo aquellos
# elementos que cumplan con la condición.
```

```python
# Dada la lista: [1, 2, 3, 4, 5]
# La comprensión de lista debería producir: [1, 9, 25]

# Lista original de números
numeros = [1, 2, 3, 4, 5]

# Comprensión de lista para obtener los cuadrados de los números impares
cuadrados_impares = [n**2 for n in numeros if n % 2 != 0]

print(cuadrados_impares)

```

```python
# El bucle while es útil cuando no se sabe de antemano cuántas veces se necesita
# ejecutar el bucle, a diferencia de un bucle for, que se utiliza cuando se conoce
#  el número de iteraciones o se itera sobre una colección de elementos.

# El bucle while en Python es una estructura de control de flujo que permite
# repetir la ejecución de un bloque de código mientras una condición específica
# sea verdadera. En cada iteración del bucle, la condición es evaluada; si la
# condición es verdadera, el código dentro del bucle se ejecuta. Si la condición
# es falsa, el bucle deja de ejecutarse y el flujo del programa continúa con
# cualquier código que esté después del bucle.

# Aquí tienes un resumen de su funcionamiento:

# Condición inicial: Antes de entrar al bucle, se establece una condición inicial
# que en principio es verdadera.
# Evaluación de la condición: El bucle while evalúa la condición; si es verdadera,
# se ejecuta el bloque de código.
# Ejecución del bloque de código: Se ejecuta todo el código dentro del bloque del
# bucle while.
# Actualización de la condición: Generalmente, dentro del bucle, algo debe cambiar
# para que eventualmente la condición se vuelva falsa. Si no hay nada que cambie
# la condición, podría resultar en un bucle infinito.
# Repetición o terminación: Después de ejecutar el bloque de código, la condición
# se evalúa nuevamente. Si sigue siendo verdadera, el bucle se repite. Si es
# falsa, el bucle termina.



```

# 14.- FUNCIONES


Las funciones en Python son bloques de código reutilizables diseñados para realizar una tarea específica. Sirven para organizar y modularizar el código, lo que facilita su lectura, mantenimiento y depuración. Al utilizar funciones, puedes dividir un programa complejo en partes más pequeñas y manejables.

Características y usos de las funciones en Python:

- Definición y Llamada:

Se definen usando la palabra clave def, seguida de un nombre, paréntesis y dos puntos. Por ejemplo: def mi_funcion():.
Se llaman (o ejecutan) usando su nombre seguido de paréntesis. Por ejemplo: mi_funcion().

- Parámetros y Argumentos:

Las funciones pueden recibir parámetros, que son variables que se utilizan dentro de la función. Los parámetros permiten pasar datos a la función.
Los argumentos son los valores reales que se pasan a los parámetros de la función cuando se llama. Por ejemplo, en mi_funcion(param1, param2), param1 y param2 son parámetros.

- Return:
Las funciones pueden devolver valores usando la palabra clave return. Esto permite que la función envíe un resultado que puede ser utilizado en otra parte del programa.
Si una función no tiene una instrucción return, o si la instrucción return no tiene un valor asociado, la función devuelve None por defecto.
Modularidad y Reutilización:

- Las funciones permiten reutilizar código. En lugar de repetir el mismo código en diferentes lugares, puedes definir una función y llamarla cada vez que necesites realizar esa tarea específica.
Esto también facilita la actualización y el mantenimiento del código, ya que los cambios solo necesitan realizarse en un lugar.


- Ejemplos de Uso:

Funciones para realizar cálculos matemáticos, como calcular el área de una figura.
Funciones para procesar datos, como manipular cadenas de texto o listas.
Funciones para interactuar con el usuario, como obtener la entrada del usuario y mostrar resultados.

#15.- CLASES Y OBJETOS

- Clases:

Una clase es como un "blueprint" o plantilla para crear objetos. Define un tipo de datos en términos de variables (atributos) y funciones (métodos) que están relacionados y se pueden utilizar para crear instancias de ese tipo.
Las clases permiten encapsular datos y las operaciones que se pueden realizar con esos datos. Esto facilita la gestión de la complejidad en programas grandes, ya que puedes pensar en términos de tipos de datos personalizados y su comportamiento asociado.

- Objetos:

Un objeto es una instancia de una clase. Cuando se crea un objeto, se está utilizando la plantilla definida por la clase para crear una entidad independiente con sus propios datos y comportamiento.
Cada objeto tiene su propio conjunto de atributos (que pueden tener valores diferentes de otros objetos de la misma clase) y métodos (que son las mismas operaciones definidas en la clase, pero pueden actuar sobre los datos del objeto).
